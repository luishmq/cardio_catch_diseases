## 0.1 Helper Functions

```python
def jupyter_settings():
    %matplotlib inline
    %pylab inline

    plt.style.use( 'bmh' )
    plt.rcParams['figure.figsize'] = [25, 12]
    plt.rcParams['font.size'] = 24
    
    display( HTML( '<style>.container { width:100% !important; }</style>') )
    pd.options.display.max_columns = None
    pd.options.display.max_rows = None
    pd.set_option( 'display.expand_frame_repr', False )
    
    sns.set()

def ml_error( model_name, y, yhat ):
    acc = accuracy_score( y, yhat )
    f1 = f1_score( y, yhat )
    recall = recall_score(y, yhat) 
    precision = precision_score(y, yhat) 
    
    return pd.DataFrame( { 'Model Name': model_name,
                            'Accuracy': acc,
                            'f1_score': f1,
                            'Recall': recall,
                            'Precision': precision}, index=[0] )
```

# 5.0 Data Preparation

```python
df5 = df4.copy()
```

```python
X = df5.drop('cardio', axis=1)
y = df5['cardio'].copy()

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

df5 = pd.concat([X_train, y_train], axis=1)
```

```python
df5.head()
```

```python
mms_lo = MinMaxScaler()
mms_hi = MinMaxScaler()
mms_we = MinMaxScaler()
mms_he = MinMaxScaler()
mms_ag = MinMaxScaler()

# age
X_train['age'] = mms_ag.fit_transform( X_train[['age']].values )

# height
X_train['height'] = mms_he.fit_transform(X_train[['height']].values )

# weight
X_train['weight'] = mms_we.fit_transform( X_train[['weight']].values )

# ap_hi
X_train['ap_hi'] = mms_hi.fit_transform( X_train[['ap_hi']].values )

# ap_lo
X_train['ap_lo'] = mms_lo.fit_transform( X_train[['ap_lo']].values )
```

```python
# age
X_test['age'] = mms_ag.transform( X_test[['age']].values )

# height
X_test['height'] = mms_he.transform(X_test[['height']].values )

# weight
X_test['weight'] = mms_we.transform( X_test[['weight']].values )

# ap_hi
X_test['ap_hi'] = mms_hi.transform( X_test[['ap_hi']].values )

# ap_lo
X_test['ap_lo'] = mms_lo.transform( X_test[['ap_lo']].values )
```

```python
# Training model to rank features

forest = en.ExtraTreesClassifier(n_estimators=250, random_state=42, n_jobs=-1)
x_train_f = X_train
y_train_f = y_train.values
forest.fit(x_train_f, y_train_f)

# Collect features from model selector
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# Organize and print ranking per features
print('Feature ranking:')
df = pd.DataFrame()
for i, j in zip(x_train_f, forest.feature_importances_):
    aux = pd.DataFrame({'feature': i, 'importance': j}, index=[0])
    df = pd.concat([df, aux], axis=0)

print(df.sort_values('importance', ascending=False))

# plot the impurity-based feature importances of the forest
plt.figure()
plt.title('Feature importances')
plt.bar(range(x_train_f.shape[1]), importances[indices], color='r', yerr=std[indices], align='center')
plt.xticks(range(x_train_f.shape[1]), indices)
plt.xlim([-1, x_train_f.shape[1]])
plt.show()
```

# 6.0 Machine Learning Modelling

```python
logr = LogisticRegression()

model = logr.fit(X_train, y_train)

yhat_logr = model.predict(X_test)

# performance
lgr_result = ml_error('Logistic Regression', y_test, yhat_logr)
lgr_result
```

```python
rfc = RandomForestClassifier(max_depth=9, n_estimators=600, random_state=42)
rfc.fit(X_train, y_train)

yhat_rfc = rfc.predict( X_test )

# performance
rfc_result = ml_error('Random Forest Classifier', y_test, yhat_rfc)
```
